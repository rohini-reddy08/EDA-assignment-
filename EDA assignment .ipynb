```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\rohini\Downloads\archive\House Price India.csv") 
```

```python
# Display the first 13 rows
print(df.head())
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Dataset summary: data types, number of rows, and columns
print("\nData Types:\n", df.dtypes)
print(f"\nNumber of Rows: {df.shape[0]}, Number of Columns: {df.shape[1]}")
```

```python
# Descriptive statistics of numerical columns
print("\nDescriptive Statistics:\n", df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'numerical_column' with a column name from your dataset
sns.histplot(df['number of floors'], kde=True)
plt.title('Distribution of number of floors')
plt.show()
```

```python
# Correlation heatmap
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Replace 'categorical_column' with a column name from your dataset
sns.countplot(x='condition of the house', data=df)
plt.title('Count Plot of condition of the house')
plt.show()
```
